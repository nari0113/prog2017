```python
import random
rand_num = 100
while rand_num != 4:
    rand_num = random.randint(1,100)
    print(rand_num)
```

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        break
    print('.')
```

breakして無限ループを止める。

```python
import random

while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
    print('.')
```

continueは4が出てもまた無限ループが始まる。
4の後は[.]は出ない。

↓問114の正解

```python
for x in range(1,21):
    if x % 3 ==0:
        if x % 5 ==0:
            print('Fizz Buzz')
        else:
            print('Fizz')
    elif x % 5 ==0:
        print('Buzz')
    else:
        print(x)
```

```python
for x in range(1,21):
    if x % 3 ==0 and x % 5!= 0:
        print('Fizz')
    if x % 3!= 0 and x % 5 ==0:
        print('Buzz')
    if x % 3 ==0 and x % 5 ==0:
        print('Fizz Buzz')
    if x % 3 !=0 and x % 5 !=0:
        print(x)
```
